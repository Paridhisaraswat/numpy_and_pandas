```python
import pandas as pd
```

```python
df = pd.read_csv('/content/test_data.csv')
df
```

```python
df.dtypes
```

```python
c = []
n =[]
num ={}
cat ={}

for i in df.columns:
  if (df[i].dtypes == 'object' and i != 'Name'):
    c.append(i)
  else:
    n.append(i)
for i in range(len(c)):
  val = c[i]
  cat[val] = df[val]

categorical_df = pd.DataFrame(cat)

for i in range(len(n)):
  val = n[i]
  num[val] = df[val]

numerical_df = pd.DataFrame(num)
```

```python
categorical_df
```

```python
numerical_df
```

```python
categorical_df['Gender'] = categorical_df['Gender'].map({'Female':0,'Male':1})
categorical_df['Department'] = categorical_df['Department'].map({'Sales':0,'IT':1,'HR':2,'Finance':3})
categorical_df['Smoker'] = categorical_df['Smoker'].map({"No":0,'Yes':1})
categorical_df['ExerciseLevel']=categorical_df['ExerciseLevel'].map({'Low':0,"Medium":1,"High":2})
categorical_df['Has_Diabetes']=categorical_df['Has_Diabetes'].map({'No':0,'Yes':1})
```
